# Информация о Датасете

Данные электрокардиограммы (ЭКГ) представляют собой записи электрической активности сердца во времени. Они являются важным инструментом для диагностики сердечных заболеваний и мониторинга состояния сердечной деятельности. Наборы данных ЭКГ могут иметь различные особенности в зависимости от их источника, целей и характеристик.

1. **Разнообразие пациентов:**
   - Наборы данных могут включать записи ЭКГ от различных пациентов, в том числе детей, взрослых и пожилых людей.
   - Могут быть представлены данные от здоровых людей и от тех, у кого есть различные сердечные заболевания.

2. **Формат данных:**
   - Данные ЭКГ могут представляться в различных форматах, таких как текстовые файлы, изображения или временные ряды.
   - Временные ряды обычно представляют собой последовательность отсчетов электрической активности сердца в течение определенного времени.

3. **Разрешение:**
   - Разрешение временных рядов может различаться в зависимости от источника данных. Например, запись ЭКГ может производиться с высокой частотой дискретизации для более детального анализа.

4. **Классификация:**
   - Разметка является ключевым фактором в использовании того или иного набора данных для решения задачи. Так, например, в электрокардиограмме могут быть размечены начальные, пиковые и конечные маркеры зубцов ЭКГ

   P - Маркер P обозначает начало и конец P-волны. P-волна представляет собой деполяризацию предсердий, что указывает на сокращение предсердий.

   PR - PR-интервал измеряет временной интервал от конца P-волны до начала QRS-комплекса. Этот интервал включает в себя время проведения импульса через атриовентрикулярный узел.
   
   QRS - QRS-комплекс представляет собой деполяризацию желудочков, что означает сокращение желудочков. Маркеры Q, R и S обозначают соответственно начало, пик и конец этого комплекса.
   
   ST - ST-сегмент начинается после QRS-комплекса и предшествует T-волне. Он представляет собой период времени между завершением желудочковой деполяризации и началом реполяризации.
   
   T - T-волна представляет собой реполяризацию желудочков. Маркер T обозначает начало и конец этой волны.
   
   U - U-волна является маленькой волной, которая иногда видна после T-волны. Она представляет собой дополнительную реполяризацию желудочков.
   
   разметка самого сердечного удара
   
   Normal - Сердечный сокращение считается нормальным, без аномалий.
   
   PVC - Раннее желудочковое сокращение, когда один из желудочков сокращается раньше обычного.
   
   PAB - Сокращение, вызванное электрической стимуляцией сердца (искусственный пульс).
   
   RBB - Блокада правой ножки пучка Гиса, что может влиять на проведение электрических сигналов в правом желудочке.
   
   LBB - Блокада левой ножки пучка Гиса, что может влиять на проведение электрических сигналов в левом желудочке.
   
   APB - Раннее предсердное сокращение.

   AFW - Маркер атриальной мерцательной волны, что может быть признаком атриальной мерцательной аритмии.
   
   VEB - Желудочковый удар, который происходит, когда сокращение не было инициировано предсердием.
   
   - Это позволяет использовать методы машинного обучения для автоматической классификации.

5. **Длительность записей:**
   - Бывают краткосрочные электрокардиограммы (30-60 секунд) и долгосрочные. Краткосрочных данных намного больше, поскольку их проще собирать, например, краткосрочная ЭКГ является основным диагностическим инструментом в амбулаторных отделениях, но даже на них могут быть обнаружены потенциальные заболевания. Однако, их может быть недостаточно для заболеваний сердца, которые нечасто проявляют себя, например, фибрилляция желудочков. ¬

6. **Шум и артефакты:**
   - Данные ЭКГ могут содержать шум, вызванный различными факторами, такими как движение пациента, электромагнитные помехи и другие артефакты.
   - Некоторые наборы данных включают в себя артефакты, чтобы сделать задачу обучения более сложной.

7. **Дополнительные параметры:**
   - Некоторые наборы данных могут включать в себя дополнительные параметры, такие как возраст, пол, медицинская история пациента, что может быть полезным для более широкого анализа.

Примеры известных наборов данных ЭКГ включают MIT-BIH Arrhythmia Database, PhysioNet databases, и PTB Diagnostic ECG Database.


MIT-BIH Arrhythmia Database является весьма известным и широко применяемым набором данных ЭКГ в мире. Его значимость обусловлена рядом ключевых характеристик:

1. **Разнообразие аритмий:**
   Набор данных включает записи ЭКГ с разнообразными типами аритмий, охватывая как нормальную электрическую активность сердца, так и различные аритмии, такие как предсердные и желудочковые экстрасистолы, фибрилляция предсердий и другие.

2. **Реальные клинические данные:**
   База данных MIT-BIH Arrhythmia Database содержит 48 получасовых фрагментов двухканальных амбулаторных записей ЭКГ, полученных от 47 человек, исследованных в лаборатории аритмий BIH в период с 1975 по 1979 год.

3. **Доступность:**
   Набор данных входит в состав открытого ресурса PhysioNet, предоставляя бесплатный доступ к медицинским данным и инструментам для исследований в области биомедицинской инженерии и физиологии.

4. **Широкое использование в исследованиях:**
   MIT-BIH Arrhythmia Database активно применяется в научных исследованиях и тестировании новых методов анализа ЭКГ. Это делает его ценным для сравнения результатов различных исследований и методов.

5. **Разметка данных:**
   Каждая запись в наборе данных размечена экспертами, что упрощает процесс обучения и проверки моделей машинного обучения на задачах классификации аритмий.

6. **Длительные записи:**
   Важной особенностью является наличие длительных записей, позволяющих анализировать долгосрочные изменения в активности сердца и выявлять аритмии в различных условиях.

7. **История использования:**
   Представленный в 1980 году, MIT-BIH Arrhythmia Database обладает долгой историей использования, свидетельствующей о его надежности и ценности для исследовательского и врачебного сообщества.

Эти факторы превращают MIT-BIH Arrhythmia Database в предпочтительный выбор для тестирования и разработки новых методов анализа ЭКГ, а также для обучения и валидации моделей машинного обучения в области кардиологии.
```python
import os
import cv2
import matplotlib.pyplot as plt
import numpy as np
import wfdb
from sklearn.preprocessing import scale
from wfdb import rdrecord

# Choose from peak to peak or centered
# mode = [20, 20]
mode = 128

image_size = 128
output_dir = '/content/data'  # Assuming the data is in the mitdb folder in Colab

# dpi fix
fig = plt.figure(frameon=False)
dpi = fig.dpi

# fig size / image size
figsize = (image_size / dpi, image_size / dpi)
image_size = (image_size, image_size)


# Assuming your data is in the /content/mitdb folder
mitdb_folder = '/content/mitdb'

for file_name in os.listdir(mitdb_folder):
    file_path = os.path.join(mitdb_folder, file_name[:-4])
    record =  rdrecord(file_path)
    ann = wfdb.rdann(file_path, extension='atr')

    for sig_name, signal in zip(record.sig_name, record.p_signal.T):
        if not np.all(np.isfinite(signal)) or sig_name != "MLII":
            continue
        signal = scale(signal)
        for i, (label, peak) in enumerate(zip(ann.symbol, ann.sample)):
            if label == '/':
                label = "\\"
            print('\r{} [{}/{}]'.format(sig_name, i + 1, len(ann.symbol)), end="")
            if isinstance(mode, list):
                if np.all([i > 0, i + 1 < len(ann.sample)]):
                    left = ann.sample[i - 1] + mode[0]
                    right = ann.sample[i + 1] - mode[1]
                else:
                    continue
            elif isinstance(mode, int):
                left, right = peak - mode // 2, peak + mode // 2
            else:
                raise Exception("Wrong mode in script beginning")

            if np.all([left > 0, right < len(signal)]):
                one_dim_data_dir = os.path.join(output_dir, '1D', file_name, sig_name, label)
                os.makedirs(one_dim_data_dir, exist_ok=True)

                filename = os.path.join(one_dim_data_dir, '{}.npy'.format(peak))
                np.save(filename, signal[left:right])

```

```python
import json
import os.path as osp
from glob import glob

import pandas as pd
# 1. N - Normal
# 2. V - PVC (Premature ventricular contraction)
# 3. \ - PAB (Paced beat)
# 4. R - RBB (Right bundle branch)
# 5. L - LBB (Left bundle branch)
# 6. A - APB (Atrial premature beat)
# 7. ! - AFW (Ventricular flutter wave)
# 8. E - VEB (Ventricular escape beat)

classes = ['N', 'V', '\\', 'R', 'L', 'A', '!', 'E']
lead = 'MLII'
extension = 'npy'  # or `npy` for 1D
data_path = osp.abspath('./data/*/*/*/*/*.{}'.format(extension))
val_size = 0.1  # [0, 1]

output_path = '/'.join(data_path.split('/')[:-5])
random_state = 7


dataset = []
files = glob(data_path)

for file in glob(data_path):
    *_, name, lead, label, filename = file.split('/')
    if lead == 'MLII':
      dataset.append({
            "name": name,
            "lead": lead,
            "label": label,
            "filename": osp.splitext(filename)[0],
            "path": file
        })

      data = pd.DataFrame(dataset)
      data = data[data['lead'] == lead]
      data = data[data['label'].isin(classes)]
      data = data.sample(frac=1, random_state=random_state)

      val_ids = []
      for cl in classes:
        val_ids.extend(data[data['label'] == cl].sample(frac=val_size, random_state=random_state).index)

      val = data.loc[val_ids, :]
      train = data[~data.index.isin(val.index)]

      train.to_json(osp.join(output_path, 'train.json'), orient='records')
      val.to_json(osp.join(output_path, 'val.json'), orient='records')

      d = {}
      for label in train.label.unique():
          d[label] = len(d)

      with open(osp.join(output_path, 'class-mapper.json'), 'w') as file:
        file.write(json.dumps(d, indent=1))
```
